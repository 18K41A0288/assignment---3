4.	Write a NumPy program to compute the condition number of a given matrix..	

Original matrix:

[[1 2]

[3 4]]

Condition number of the said matrix:
	14.9330343737


```
import numpy as np
g = np.array([[1,2],[3,4]])
print("Condition number of the said matrix : ",np.linalg.cond(g))
```
